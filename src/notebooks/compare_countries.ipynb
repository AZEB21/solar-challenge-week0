```python
import pandas as pd
import matplotlib.pyplot as plt

benin = pd.read_csv("../../data/benin_clean.csv")
sierra = pd.read_csv("../../data/sierra_leone_clean.csv")
togo = pd.read_csv("../../data/togo_clean.csv")

# Add country column
benin["Country"] = "Benin"
sierra["Country"] = "Sierra Leone"
togo["Country"] = "Togo"

# Combine datasets
df = pd.concat([benin, sierra, togo], ignore_index=True)

# Summary statistics
df.groupby("Country")[["GHI","DNI","DHI"]].describe()
```

```python
plt.figure(figsize=(10,6))
df.boxplot(column="GHI", by="Country")
plt.title("GHI Comparison Across Countries")
plt.suptitle("")
plt.show()
```

```python
df.groupby("Country")["GHI"].mean().plot(kind="bar", figsize=(7,5), title="Average GHI by Country")
plt.show()
```
